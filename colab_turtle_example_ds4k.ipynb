```
from ColabTurtle.Turtle import * # Module/Package
turtle_speed = 7
initializeTurtle(initial_speed=turtle_speed)
t = ColabTurtle.Turtle # Create a turtle
t.pen_width = 1
colors = ["red", "yellow","blue","green"]
for x in range(100):
  t.pen_color = colors[x%4] # Loop over "red", "yellow","blue","green"
  t.forward(x) # Move forward x pixels 0-99
  t.left(99) # Turn left 91 degrees
```
